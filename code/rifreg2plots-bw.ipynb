```python
wmen8388 = wrifdata8388[wrifdata8388.name.str.startswith("m")]
wwom8388 = wrifdata8388[wrifdata8388.name.str.startswith("f")]

wmen8800 = wrifdata8800[wrifdata8800.name.str.startswith("m")]
wwom8800 = wrifdata8800[wrifdata8800.name.str.startswith("f")]

wmen0019 = wrifdata0019[wrifdata0019.name.str.startswith("m")]
wwom0019 = wrifdata0019[wrifdata0019.name.str.startswith("f")]

bmen8388 = brifdata8388[brifdata8388.name.str.startswith("m")]
bwom8388 = brifdata8388[brifdata8388.name.str.startswith("f")]

bmen8800 = brifdata8800[brifdata8800.name.str.startswith("m")]
bwom8800 = brifdata8800[brifdata8800.name.str.startswith("f")]

bmen0019 = brifdata0019[brifdata0019.name.str.startswith("m")]
bwom0019 = brifdata0019[brifdata0019.name.str.startswith("f")]
```

# White Men

```python
#Association between coverage and wages for all white men along distribution 8388
coveredwmen8388 = wmen8388[wmen8388.name.str.startswith("m_covered")]
coveredwmen8388["percentile"] = coveredwmen8388["name"].str.strip().str[-2:].astype(int)

coveredwmen8388["ols"] = 0
coveredwmen8388["lbols"] = 0
coveredwmen8388["ubols"] = 0

fig, ax = plt.subplots()
ax.axhline(y = 0, color = "black", linestyle ="--")
ax.plot(coveredwmen8388.percentile, coveredwmen8388.coef)
ax.fill_between(coveredwmen8388.percentile, coveredwmen8388.lb, coveredwmen8388.ub, alpha = 0.1)
ax.plot(coveredwmen8388.percentile, coveredwmen8388.ols, color = "orange")
ax.fill_between(coveredwmen8388.percentile, coveredwmen8388.lbols, coveredwmen8388.ubols, alpha = 0.1, color = "orange")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wrifdiddata = pd.read_excel("/Users/amedeusdsouza/econ499/linreg/rifdid_coefs-sltt_blk0.xlsx")
brifdiddata = pd.read_excel("/Users/amedeusdsouza/econ499/linreg/rifdid_coefs-sltt_blk1.xlsx")
```

```python
wmen = wrifdiddata[wrifdiddata.name.str.startswith("m")]
wumen = wrifdiddata[wrifdiddata.name.str.startswith("um")]
wnmen = wrifdiddata[wrifdiddata.name.str.startswith("nm")]

wwom = wrifdiddata[wrifdiddata.name.str.startswith("f")]
wuwom = wrifdiddata[wrifdiddata.name.str.startswith("uf")]
wnwom = wrifdiddata[wrifdiddata.name.str.startswith("nf")]

bmen = brifdiddata[brifdiddata.name.str.startswith("m")]
bumen = brifdiddata[brifdiddata.name.str.startswith("um")]
bnmen = brifdiddata[brifdiddata.name.str.startswith("nm")]

bwom = brifdiddata[brifdiddata.name.str.startswith("f")]
buwom = brifdiddata[brifdiddata.name.str.startswith("uf")] #note maybe collinearity due to lack of obs FIXME
bnwom = brifdiddata[brifdiddata.name.str.startswith("nf")]
```

# ALL MEN
# White Men

```python
#Effect of treat_st on wages for all white men along distribution
whitemen = wmen[wmen.name.str.startswith("m_treat_st")]
whitemen["percentile"] = whitemen["name"].str.strip().str[-2:].astype(int)

whitemen["ols"] = 0
whitemen["lbols"] = 0
whitemen["ubols"] = 0

fig, ax = plt.subplots()
ax.axhline(y = 0, color = "black", linestyle ="--")
ax.plot(whitemen.percentile, whitemen.coef)
ax.fill_between(whitemen.percentile, whitemen.lb, whitemen.ub, alpha = 0.1)
ax.plot(whitemen.percentile, whitemen.ols, color = "orange")
ax.fill_between(whitemen.percentile, whitemen.lbols, whitemen.ubols, alpha = 0.1, color = "orange")
```

# Black Men

```python
#Effect of treat_st on wages for all Black men along distribution
blackmen = bmen[bmen.name.str.startswith("m_treat_st")]
blackmen["percentile"] = blackmen["name"].str.strip().str[-2:].astype(int)

blackmen["ols"] = 0
blackmen["lbols"] = 0
blackmen["ubols"] = 0

fig, ax = plt.subplots()
ax.axhline(y = 0, color = "black", linestyle ="--")
ax.plot(blackmen.percentile, blackmen.coef, color = "darkviolet")
ax.fill_between(blackmen.percentile, blackmen.lb, blackmen.ub, alpha = 0.1, color = "hotpink")
ax.plot(blackmen.percentile, blackmen.ols, color = "orange")
ax.fill_between(blackmen.percentile, blackmen.lbols, blackmen.ubols, alpha = 0.1, color = "orange")
```

```python
# Table RIF-DiD Regression of Real Log Wages on Right to Work Laws Treatment in State and Time - for All (B/w) People Separately - with SLTT

fig, ax = plt.subplots(2,2, figsize = (35, 20), sharey = True, sharex = True)

for i in [0,1]:
    for j in [0,1]:
        ax[i,j].set_yticks(np.arange(-0.5, 0.5, step=0.05))
        ax[i,j].xaxis.set_tick_params(labelbottom=True)
        ax[i,j].yaxis.set_tick_params(labelbottom=True)
#         ax[i,j].set_xlabel("Percentiles for All (Black and White) Men")
        ax[i,j].set_ylabel("Real Log Wages")

ax[0,0].set_xlabel("Percentiles for All White Men")
ax[1,0].set_xlabel("Percentiles for All White Women")
ax[0,1].set_xlabel("Percentiles for All Black Men")
ax[1,1].set_xlabel("Percentiles for All Black Women")

#White Men - Treat_st
ax[0,0].axhline(y = 0, color = "black", linestyle ="--")
ax[0,0].plot(whitemen.percentile, whitemen.coef)
ax[0,0].fill_between(whitemen.percentile, whitemen.lb, whitemen.ub, alpha = 0.1)
ax[0,0].plot(whitemen.percentile, whitemen.ols, color = "orange")
ax[0,0].fill_between(whitemen.percentile, whitemen.lbols, whitemen.ubols, alpha = 0.1, color = "orange")

#White Women - Treat_st
ax[1,0].axhline(y = 0, color = "black", linestyle ="--")
ax[1,0].plot(whitewom.percentile, whitewom.coef)
ax[1,0].fill_between(whitewom.percentile, whitewom.lb, whitewom.ub, alpha = 0.1)
ax[1,0].plot(whitewom.percentile, whitewom.ols, color = "orange")
ax[1,0].fill_between(whitewom.percentile, whitewom.lbols, whitewom.ubols, alpha = 0.1, color = "orange")

#Black Men - Treat_st
ax[0,1].axhline(y = 0, color = "black", linestyle ="--")
ax[0,1].plot(blackmen.percentile, blackmen.coef, color = "darkviolet")
ax[0,1].fill_between(blackmen.percentile, blackmen.lb, blackmen.ub, alpha = 0.1, color = "hotpink")
ax[0,1].plot(blackmen.percentile, blackmen.ols, color = "orange")
ax[0,1].fill_between(blackmen.percentile, blackmen.lbols, blackmen.ubols, alpha = 0.1, color = "orange")

#Black Women - Treat_st
ax[1,1].axhline(y = 0, color = "black", linestyle ="--")
ax[1,1].plot(blackwom.percentile, blackwom.coef, color = "darkviolet")
ax[1,1].fill_between(blackwom.percentile, blackwom.lb, blackwom.ub, alpha = 0.1, color = "hotpink")
ax[1,1].plot(blackwom.percentile, blackwom.ols, color = "orange")
ax[1,1].fill_between(blackwom.percentile, blackwom.lbols, blackwom.ubols, alpha = 0.1, color = "orange")

cols = ['{}'.format(col) for col in ["White", "Black"]]
rows = ['{}'.format(row) for row in ["Treat_st", "Treat_st"]]

pad = 5 # in points

for a, col in zip(ax[0], cols):
    a.annotate(col, xy=(0.5, 1), xytext=(0, pad),
                xycoords='axes fraction', textcoords='offset points',
                size=27, ha='center', va='baseline')

for a, row in zip(ax[:,0], rows):
    a.annotate(row, xy=(0, 0.5), xytext=(-a.yaxis.labelpad - pad, 0),
                xycoords=a.yaxis.label, textcoords='offset points',
                size=27, ha='right', va='center')


fig.subplots_adjust(left=0.15, top=0.95)

plt.show()
```

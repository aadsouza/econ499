```python
# Table RIF-DiD Regression of Real Log Wages on Right to Work Laws Treatment in State and Time - for All (B/w) People - with SLTT
fig, ax = plt.subplots(3,2, figsize = (35, 20), sharey = True, sharex = True)

for i in [0,1,2]:
    for j in [0,1]:
        ax[i,j].set_yticks(np.arange(-0.5, 0.5, step=0.05))
        ax[i,j].xaxis.set_tick_params(labelbottom=True)
        ax[i,j].yaxis.set_tick_params(labelbottom=True)
#         ax[i,j].set_xlabel("Percentiles for All (Black and White) Men")
        ax[i,0].set_xlabel("Percentiles for All (Black and White) Men")
        ax[i,1].set_xlabel("Percentiles for All (Black and White) Women")
        ax[i,j].set_ylabel("Real Log Wages")

#Men - Black
ax[0,0].axhline(y = 0, color = "black", linestyle ="--")
ax[0,0].plot(blackmen.percentile, blackmen.coef, color = "darkviolet")
ax[0,0].fill_between(blackmen.percentile, blackmen.lb, blackmen.ub, color = "hotpink", alpha = 0.1)
ax[0,0].plot(blackmen.percentile, blackmen.ols, color = "orange")
ax[0,0].fill_between(blackmen.percentile, blackmen.lbols, blackmen.ubols, alpha = 0.1, color = "orange")

#Men - Treat
ax[1,0].axhline(y = 0, color = "black", linestyle ="--")
ax[1,0].plot(whitemen.percentile, whitemen.coef)
ax[1,0].fill_between(whitemen.percentile, whitemen.lb, whitemen.ub, alpha = 0.1)
ax[1,0].plot(whitemen.percentile, whitemen.ols, color = "orange")
ax[1,0].fill_between(whitemen.percentile, whitemen.lbols, whitemen.ubols, alpha = 0.1, color = "orange")

#Men - Times
ax[2,0].axhline(y = 0, color = "black", linestyle ="--")
ax[2,0].plot(blackmen_treat.percentile, blackmen_treat.coef, color = "darkviolet")
ax[2,0].fill_between(blackmen_treat.percentile, blackmen_treat.lb, blackmen_treat.ub, color = "hotpink", alpha = 0.1)
ax[2,0].plot(blackmen_treat.percentile, blackmen_treat.ols, color = "orange")
ax[2,0].fill_between(blackmen_treat.percentile, blackmen_treat.lbols, blackmen_treat.ubols, alpha = 0.1, color = "orange")

#Women - Black
ax[0,1].axhline(y = 0, color = "black", linestyle ="--")
ax[0,1].plot(blackwom.percentile, blackwom.coef, color = "darkviolet")
ax[0,1].fill_between(blackwom.percentile, blackwom.lb, blackwom.ub, color = "hotpink", alpha = 0.1)
ax[0,1].plot(blackwom.percentile, blackwom.ols, color = "orange")
ax[0,1].fill_between(blackwom.percentile, blackwom.lbols, blackwom.ubols, alpha = 0.1, color = "orange")

#Women - Treat
ax[1,1].axhline(y = 0, color = "black", linestyle ="--")
ax[1,1].plot(whitewom.percentile, whitewom.coef)
ax[1,1].fill_between(whitewom.percentile, whitewom.lb, whitewom.ub, alpha = 0.1)
ax[1,1].plot(whitewom.percentile, whitewom.ols, color = "orange")
ax[1,1].fill_between(whitewom.percentile, whitewom.lbols, whitewom.ubols, alpha = 0.1, color = "orange")

#Women - Times
ax[2,1].axhline(y = 0, color = "black", linestyle ="--")
ax[2,1].plot(blackwom_treat.percentile, blackwom_treat.coef, color = "darkviolet")
ax[2,1].fill_between(blackwom_treat.percentile, blackwom_treat.lb, blackwom_treat.ub, color = "hotpink", alpha = 0.1)
ax[2,1].plot(blackwom_treat.percentile, blackwom_treat.ols, color = "orange")
ax[2,1].fill_between(blackwom_treat.percentile, blackwom_treat.lbols, blackwom_treat.ubols, alpha = 0.1, color = "orange")

cols = ['{}'.format(col) for col in ["Men", "Women"]]
rows = ['{}'.format(row) for row in ["Black", "Treat_st", "Treat_st * Black"]]

pad = 5 # in points

for a, col in zip(ax[0], cols):
    a.annotate(col, xy=(0.5, 1), xytext=(0, pad),
                xycoords='axes fraction', textcoords='offset points',
                size=27, ha='center', va='baseline')

for a, row in zip(ax[:,0], rows):
    a.annotate(row, xy=(0, 0.5), xytext=(-a.yaxis.labelpad - pad, 0),
                xycoords=a.yaxis.label, textcoords='offset points',
                size=27, ha='right', va='center')


fig.subplots_adjust(left=0.15, top=0.95)

plt.show()
```

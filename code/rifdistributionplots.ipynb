```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rifdiddata = pd.read_excel("/Users/amedeusdsouza/econ499/linreg/rifdid_coefs.xlsx")
```

```python
rifdiddata
```

```python
men = rifdiddata[rifdiddata.name.str.startswith("m")]
umen = rifdiddata[rifdiddata.name.str.startswith("um")]
nmen = rifdiddata[rifdiddata.name.str.startswith("nm")]

wom = rifdiddata[rifdiddata.name.str.startswith("f")]
uwom = rifdiddata[rifdiddata.name.str.startswith("uf")]
nwom = rifdiddata[rifdiddata.name.str.startswith("nf")]
```

# ALL MEN AND WOMEN

```python
#Effect of treat_st on wages for all white men along distribution
whitemen = men[men.name.str.startswith("m_treat_st")]
whitemen["percentile"] = whitemen["name"].str.strip().str[-2:].astype(int)

fig, ax = plt.subplots()
ax.plot(whitemen.percentile, whitemen.coef)
ax.fill_between(whitemen.percentile, whitemen.lb, whitemen.ub, alpha = 0.1)
```

```python
#Black wage penalty for all men along distribution
blackmen = men[men.name.str.startswith("m_blackrelwhite")]
blackmen["percentile"] = blackmen["name"].str.strip().str[-2:].astype(int)

fig, ax = plt.subplots()
ax.plot(blackmen.percentile, blackmen.coef, color = "darkviolet")
ax.fill_between(blackmen.percentile, blackmen.lb, blackmen.ub, color = "hotpink", alpha = 0.1)
```

```python
#Effect of treat_stblackrelwhite for all men along distribution
blackmen_treat = men[men.name.str.startswith("m_times")]
blackmen_treat["percentile"] = blackmen_treat["name"].str.strip().str[-2:].astype(int)

fig, ax = plt.subplots()
ax.plot(blackmen_treat.percentile, blackmen_treat.coef, color = "darkviolet")
ax.fill_between(blackmen_treat.percentile, blackmen_treat.lb, blackmen_treat.ub, color = "hotpink", alpha = 0.1)
```

```python
#Effect of treat_st on wages for all white women along distribution
whitewom = wom[wom.name.str.startswith("f_treat_st")]
whitewom["percentile"] = whitewom["name"].str.strip().str[-2:].astype(int)

fig, ax = plt.subplots()
ax.plot(whitewom.percentile, whitewom.coef)
ax.fill_between(whitewom.percentile, whitewom.lb, whitewom.ub, alpha = 0.1)
```
